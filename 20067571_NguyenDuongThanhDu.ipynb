4. Chia bộ dữ liệu thành 2 phần Train/Test với các tỉ lệ nhất định

```python
from sklearn.model_selection import train_test_split

X = df.drop('class', axis=1)
#drop attribute 6 because it has the lowest correlation with other attributes
X = X.drop('attribute 6', axis=1)
y = df['class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=13)
```

5. Đánh giá bộ dữ liệu với giải thuật KNN bằng phương pháp 10-Fold (k-fold)

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
KFold = KFold(n_splits=10, random_state=13, shuffle=True)
knn = KNeighborsClassifier()
#grid search to find the best k
from sklearn.model_selection import GridSearchCV
param_grid = {'n_neighbors': np.arange(1, 30)}
knn_gscv = GridSearchCV(knn, param_grid, cv=KFold)
knn_gscv.fit(X_train, y_train)
knn_gscv.best_params_


```

```python
from sklearn.model_selection import cross_val_score
knn = KNeighborsClassifier(n_neighbors=knn_gscv.best_params_['n_neighbors'])
# knn = KNeighborsClassifier(n_neighbors=)
cv_scores = cross_val_score(knn, X, y, cv=KFold)

print(cv_scores)
print('cv_scores mean:{}'.format(np.mean(cv_scores)))
knn.fit(X_train, y_train)
```

6. Huấn luyện dữ liệu cho bài toán phân lớp sử dụng với các giải thuật Decision Tree,
Bayes, SVM và XGBoot.

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(random_state=13)
tree.fit(X_train, y_train)
y_pred = tree.predict(X_test)
print('Accuracy: {}'.format(accuracy_score(y_test, y_pred)))
```

```python
#Bayes
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
print('Accuracy: {}'.format(accuracy_score(y_test, y_pred)))
```

```python
#SVM
from sklearn.svm import SVC

svm = SVC(random_state=13)
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
print('Accuracy: {}'.format(accuracy_score(y_test, y_pred)))
```

Vì XGBoost yêu cầu label phải bắt đầu từ 0 nên ta sử dụng label encoder để chuyển đổi label

```python
from sklearn.preprocessing import LabelEncoder
y_train = LabelEncoder().fit_transform(y_train)
y_test = LabelEncoder().fit_transform(y_test)
#count number of each class
np.unique(y_train, return_counts=True)
```

```python
#XGBoost
from xgboost import XGBClassifier

xgb = XGBClassifier(random_state=13)
xgb.fit(X_train, y_train)
y_pred = xgb.predict(X_test)
print('Accuracy: {}'.format(accuracy_score(y_test, y_pred)))
```

10. So sánh kết quả các độ đo 6,7,8 với các giải thuật học máy ở trên

```python
#F1 score for each model
f1_knn= f1_score(y_test, knn.predict(X_test), average='weighted')
print('F1 score for KNN: {}'.format(f1_knn))
f1_tree = f1_score(y_test, tree.predict(X_test), average='weighted')
print('F1 score for Decision Tree: {}'.format(f1_tree))
f1_bayes = f1_score(y_test, gnb.predict(X_test), average='weighted')
print('F1 score for Naive Bayes: {}'.format(f1_bayes))
f1_svm = f1_score(y_test, svm.predict(X_test), average='weighted')
print('F1 score for SVM: {}'.format(f1_svm))
f1_xgb = f1_score(y_test, xgb.predict(X_test), average='weighted')
print('F1 score for XGBoost: {}'.format(f1_xgb))
```

```python
#plot F1 score for each model
f1_scores = [f1_knn, f1_tree, f1_bayes, f1_svm, f1_xgb]
models = ['KNN', 'Decision Tree', 'Naive Bayes', 'SVM', 'XGBoost']
sns.barplot(x=models, y=f1_scores, palette="Set2")
```

```python
#confusion matrix for each model
cm_knn = confusion_matrix(y_test, knn.predict(X_test))
cm_tree = confusion_matrix(y_test, tree.predict(X_test))
cm_bayes = confusion_matrix(y_test, gnb.predict(X_test))
cm_svm = confusion_matrix(y_test, svm.predict(X_test))
cm_xgb = confusion_matrix(y_test, xgb.predict(X_test))

#plot confusion matrix for each model
fig, ax = plt.subplots(nrows=2,ncols=3, figsize=(18,10))
sns.heatmap(cm_knn, annot=True, cmap='viridis', linewidths=0.2, annot_kws={'size':20}, ax=ax[0,0])
sns.heatmap(cm_tree, annot=True, cmap='viridis', linewidths=0.2, annot_kws={'size':20}, ax=ax[0,1])
sns.heatmap(cm_bayes, annot=True, cmap='viridis', linewidths=0.2, annot_kws={'size':20}, ax=ax[0,2])
sns.heatmap(cm_svm, annot=True, cmap='viridis', linewidths=0.2, annot_kws={'size':20}, ax=ax[1,0])
sns.heatmap(cm_xgb, annot=True, cmap='viridis', linewidths=0.2, annot_kws={'size':20}, ax=ax[1,1])
```
